#### Task 1.
Задание:

1. Создайте Series из последовательности 15 значений, равномерно разбивающих отрезок [0, 20] (воспользуйтесь функцией linspace)
2. Определите отношение элементов полученной серии к их предыдущим элементам (\*).
3. В результате необходимо получить среднее полученного вектора, оставив в нём только те значения, которые не более чем 1.5 (\**).

Выберите из ответов тот, который максимально близок к полученному (c точки зрения абсолютной разницы).

#### Варианты ответов:
    - 1) 1.24
    + 2) 1.18
    - 3) 0.71
    - 4) 1.13


##### Пояснения:
(\*) Если было бы необходимо найти последовательность из 3-х значений,  
равномерно разбивающих отрезок [0,1], то это были бы значения [0, 0.5, 1].

(\**) Если был бы дан список элементов a = [1,2,3,12], отношения элементов к предыдущим будут равны [NaN, 2, 1.5, 4].  
А на последнем этапе в таком примере останется только [1.5] и среднее значение будет также 1.5.

````python
# Создание Series из последовательности 15 значений, равномерно разбивающих отрезок [0, 20]
series = pd.Series(np.linspace(0, 20, 15))

# Определение отношения элементов серии к их предыдущим элементам
ratios = series / series.shift(1)

# Вычисление среднего полученного вектора, оставив только те значения, которые не более чем 1.5
filtered_ratios = ratios[(ratios <= 1.5) & (ratios >= -1.5)]  # Оставляем значения не более чем 1.5 и не менее чем -1.5
mean_ratio = filtered_ratios.mean()

print("Среднее отношение элементов, удовлетворяющих условиям:", mean_ratio)
````

#### Пояснения:
(\*) Если получились ответы 3, 4.32, то в форму необходимо внести их в виде "3,4.32". То есть вещественные числа необходимо разделять точками. Не забудьте про фиксированный seed (его менять не нужно)!
 - Определенное значение seed нужно, чтобы ответы у всех выполняющих

одинаковые series).

либо выбирать данные последовательно, либо сразу по нескольким условиям,
но через оператор &. Отличие оператора and от оператора &: and - выводит
последнее проверенное значение, & - выводит пересечение значений.
Пример: s[ _  &  _ ].sum()

````python
np.random.seed(242)

# Генерируем Series из 100 значений нормально распределенной случайной величины
s = pd.Series(np.random.normal(size=100))

# Возводим каждое значение в серии в третью степень
s = s ** 3

# Увеличиваем значения индекса в три раза
s.index = s.index * 3

# 1) Сумма элементов, строго меньших 2.6, имеющих нечетные значения индекса
result_1 = s[(s < 2.6) & (s.index % 2 != 0)].sum()

# 2) Количество значений серии меньше нуля
result_2 = (s < 0).sum()

# Выводим результаты
print("{},{}".format(result_1, result_2))
````

````python
# Вычисляем медиану суммы транзакций
median_all = transactions['amount'].median()

# Медиана суммы транзакций по тем строкам, которые ни в одном из своих столбцов не содержат пустые значения
median_no_na = transactions.dropna()['amount'].median()

# Медиана суммы транзакций по строкам, отсортированным по полю amount в порядке возрастания,
# и из которых удалены дублирующиеся по столбцам [mcc_code, tr_type] строки,
# причём при удалении соответствующих дублей остаются только последние из дублирующихся строк (keep='last')
sorted_transactions = transactions.sort_values(by='amount')
unique_transactions = sorted_transactions.drop_duplicates(subset=['mcc_code', 'tr_type'], keep='last')
median_unique_sorted = unique_transactions['amount'].median()

# Находим максимальную разницу между медианами
max_difference = np.ptp([median_all, median_no_na, median_unique_sorted])

# Выводим ответ в нужном формате
print("{:.2f}".format(max_difference))
````
